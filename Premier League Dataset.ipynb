```python
df['saves'] = df['saves'].fillna(0)
df['head_clearance'] = df['head_clearance'].fillna(0)
df['total_through_ball'] = df['total_through_ball'].fillna(0)
df['backward_pass'] = df['backward_pass'].fillna(0)
df['big_chance_missed'] = df['big_chance_missed'].fillna(0)
df['dispossessed'] = df['dispossessed'].fillna(0)
```

```python
df.isnull().sum()
```

# Q1. which team has highest win 

```python
df[df['wins'].max() == df['wins']][['team' , 'wins' , 'season']]
```

# Q2. which team has lowest losses

```python
df[df['losses'].min() == df['losses']][['team' , 'losses' , 'season']]
```

# Q3. which team score more goals

```python
df[df['goals'].max() == df['goals']][['team' , 'goals','season']]
```

# Q4. total number of red cards in season 2017-2018

```python
df[df["season"].str.contains("2017-2018" , case = False)]["total_red_card"].sum()
```

# Q5. display top 5 teams having highest goals in season 2016-2017

```python
df[df["season"].str.contains("2016-2017" , case = False)].sort_values(by=['goals'], ascending = False)[['team','goals','season']].head(5)
```

# Q6. total number of goals in season 2012-2013

```python
df[df["season"].str.contains("2012-2013" , case = False)][['goals']].sum()
```
